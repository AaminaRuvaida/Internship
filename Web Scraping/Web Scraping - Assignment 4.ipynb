# 1) Scrape the details of most viewed videos on YouTube from Wikipedia.
### Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Importing Necessary Libraries

import pandas as pd
import selenium
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Edge('msedgedriver.exe')
```

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]

# Scraping rank data:
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")[:30]:
        Rank.append(i.text)
except:
    NoSuchElementException
    Rank.append('--')
    
# Scraping Name data :
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")[:30]:
        Name.append(i.text)
except:
    NoSuchElementException
    Rank.append('--')
    
# Scraping Artist data:
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")[:30]:
        Artist.append(i.text)
except:
    NoSuchElementException
    Artist.append('--')
    
# Scraping Upload_date
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")[:30]:
        Upload_date.append(i.text)
except:
    NoSuchElementException
    Upload_date.append('--')
    
    
# Scraping Views data:
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")[:30]:
        Views.append(i.text)
except:
    NoSuchElementException
    Views.append('--')
```

```python
df_youtube=pd.DataFrame()
df_youtube['Rank']=Rank
df_youtube['Name']=Name
df_youtube['Artist']=Artist
df_youtube['Upload_date']=Upload_date
df_youtube['Views']=Views
df_youtube
```

```python
df_youtube.to_csv('top_views_youtube.csv')
```

```python
import requests
```

```python
page=requests.get("https://www.naukri.com/data-science-recruiters")
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")
```

```python
scraped_name=soup.find_all('span',class_="fl ellipsis")
```

```python
# Scraping  Name
names=[]
for i in scraped_name:
    name=i.get_text()
    name=names.append(i.text)

#Scraping Designation
designation=[]
for i in soup.find_all('span',class_='ellipsis clr'):
    desig=i.get_text()
    desig=designation.append(i.text)

#Scraping Company
company=[]
for i in soup.find_all('a',class_='ellipsis'):
    com=i.get_text()
    com=company.append(i.text)
    
#Scraping skill they hire
skill=[]
for i in soup.find_all('div',class_="hireSec highlightable"):
    skil=i.get_text().replace('  ','')
    skil=skill.append(i.text)

# Scraping Location
location=[]
for i in soup.find_all('small',class_="ellipsis"):
    loc=i.get_text()
    loc=location.append(i.text)
```

```python
df_naukri=pd.DataFrame()
df_naukri['Name']=names[:40]
df_naukri['Designation']=designation[:40]
df_naukri['Company']=company[1:80:2]
df_naukri['Skills']=skill[:40]
df_naukri['Location']=location[:40]
df_naukri
```

```python
df_naukri.to_csv('datascience_recruiter.csv')
```

```python
database_name=[]
database_type=[]
task=[]
attr_type=[]
instances=[]
attributes=[]
year=[]

#Scraping Database name:
try:
    for i in driver.find_elements_by_xpath("//p[@class='normal']//b//a"):
        database_name.append(i.text)
except:
    NoSuchElementException
    database_name.append('--')
    
# Scraping the Database Type
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']//tbody//tr//td[2]//p"):
        database_type.append(i.text)
except:
    NoSuchElementException
    database_type.append('--')
    
# Scraping the Task
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']//tbody//td[3]//p"):
        task.append(i.text)
except:
    NoSuchElementException
    task.append('--')
    
# Scraping Attribute Type
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']//tbody//td[4]//p"):
        attr_type.append(i.text)
except:
    NoSuchElementException
    attr_type.append('--')
    
# Scraping No of Instances
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']//tbody//td[5]//p"):
        instances.append(i.text)
except:
    NoSuchElementExcpetion
    instances.append('--')
    
# Scraping No of Attributes
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']//tbody//td[6]//p"):
        attributes.append(i.text)
except:
    NoSuchElementException
    attributes.append('--')
    
# Scraping Year
try:
    for i in driver.find_elements_by_xpath("//table[@border='1']//tbody//td[7]//p"):
        year.append(i.text)
except:
    NoSuchElementException
    year.append('--')
```

```python
df_uci=pd.DataFrame()
df_uci['Database Name']=database_name[:200]
df_uci['Database Type']=database_type[2:401:2]
df_uci['Task']=task[1:201]
df_uci['Attribute Type']=attr_type[1:201]
df_uci['No of Instances']=instances[1:201]
df_uci['No of Attributes']=attributes[1:201]
df_uci['Year']=year[1:201]
df_uci
```
