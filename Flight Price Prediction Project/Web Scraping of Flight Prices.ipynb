# Web Scraping of Flight Price

```python
# Importing necessary Libraries

import pandas as pd
import numpy as np
import time
import requests
from selenium import webdriver
import selenium
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementClickInterceptedException, ElementNotVisibleException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import warnings
warnings.filterwarnings('ignore')
```

```python
# Connecting to webdriver

driver=webdriver.Chrome('chromedrivers.exe')
```

```python
# Entering the website url in Chrome browser
url="https://www.yatra.com/"
driver.get(url)
```

```python
# Clicking on search flight
search_flight=driver.find_element_by_xpath("/html/body/div[2]/div/section/div[1]/div/div[1]/section/div/div/div/div[1]/div[3]/div[2]/div[1]/input")
search_flight.click()
```

```python
# Location for Scraping the data of Source and Destination
location=['Chennai','Bangalore','Hyderabad','Goa','New Delhi','Mumbai']
```

```python
# Creating an empty list

Airline_name=[]
Source=[]
Destination=[]
Departure_time=[]
Arrival_time=[]
Duration=[]
Stops=[]
Location=[]
Price=[]


for x in location:
    for y in location:
        if x!=y:
            #Entering the Source Location
            source=driver.find_element_by_xpath("/html/body/section[1]/section/div/div/form/ul[2]/li[2]/div/input")
            time.sleep(2)
            source.clear()
            source.send_keys(x)
            time.sleep(2)
            
            #Entering the Destination Location
            destination=driver.find_element_by_xpath("/html/body/section[1]/section/div/div/form/ul[2]/li[3]/div/input")
            destination.click()
            time.sleep(2)
            destination.clear()
            destination.send_keys(y)
            time.sleep(2)
            
            try:
                driver.find_element_by_xpath("/html/body/section[2]/section/div[1]/div/div[1]").click()
            except NoSuchElementException:
                pass
            
            #Click on Search Flight
            try:
                srch_btn = driver.find_element_by_xpath("//button[@class='fs-14 btn-submit cursor-pointer bold']")
                wait = WebDriverWait(driver, 10)
                wait.until(EC.visibility_of(srch_btn))
                srch_btn.click()
            except NoSuchElementException:
                pass
            time.sleep(2)
            
            #Scrolling the data
            for _ in range(15):
                time.sleep(2)
                driver.execute_script("window.scrollBy(0,5000)")
                
            #Scrolling to top button
            try:
                driver.find_element_by_xpath("//div[@title='scroll to top']").click()
            except NoSuchElementException:
                pass
            
            #Fetching the name of Airline
            try:
                for i in driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span"):
                    Airline_name.append(i.text)
            except NoSuchElementException:
                Airline_name.append('--')
                
            #Fetching the Source data
            try:
                for i in driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']"):
                    Source.append(i.text)
            except NoSuchElementException:
                Source.append('--')
                
            #Fetching the Destination data
            try:
                for i in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"):
                    Destination.append(i.text)
            except NoSuchElementException:
                Destination.append('--')
                
            #Fetching the Departure time
            try:
                for i in driver.find_elements_by_xpath("//div[@class='i-b pr']"):
                    Departure_time.append(i.text)
            except NoSuchElementException:
                Deaprture_time.append('--')
                
            #Fetching the Arrival time
            try:
                for i in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[1]"):
                    Arrival_time.append(i.text)
            except NoSuchElementException:
                Arrival_time.append('--')
                
            #Fetching the Duration
            try:
                for i in driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']/p[1]"):
                    Duration.append(i.text)
            except NoSuchElementException:
                Duration.append('--')
                
            #Fetching the Stops
            try:
                for i in driver.find_elements_by_xpath("//span[@class='cursor-default']"):
                    Stops.append(i.text)
            except NoSuchElementException:
                Stops.append('--')
                
            # Fetching Locations
            try:
                for i in driver.find_elements_by_xpath("//p[@class='fs-10 font-lightgrey no-wrap city ellipsis']"):
                    Location.append(i.text)
            except NoSuchElementException:
                pass

                
            #Fetching the Price
            try:
                for i in driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']"):
                    Price.append(i.text)
            except NoSuchElementException:
                Price.append('--')
                
    
            
```

```python
df=pd.DataFrame()
df['Airlines']=Airline_name[:2700]
df['Source']=Source[:2700]
df['Destination']=Destination[:2700]
df['Departure']=Departure_time[:2700]
df['Arrival']=Arrival_time[:2700]
df['Duration']=Duration[:2700]
df['Price']=Price[:2700]
df
```

```python
df.to_csv("Yatra_flight_price.csv")
```
