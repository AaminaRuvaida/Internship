# Scrapping of Used Car from cardekho website

```python
# Importing Libraries

import selenium
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python
# Connecting to Edge webdriver
driver=webdriver.Chrome('chromedrivers.exe')
```

```python
url = "https://www.cardekho.com/"
driver.get(url)
time.sleep(3)
```

```python
#link for city
used_cars = driver.find_element_by_xpath("/html/body/div[2]/div/header/div[2]/div/div/nav/ul/li[2]/ul/li[2]/a")    
```

```python
used_cars = used_cars.get_attribute('href')
```

```python
used_cars
```

```python
driver.get(used_cars)
time.sleep(3)
```

```python
name=[]
price=[]
running_km=[]
fuel_type=[]
transmission=[]
for _ in range(150):
        time.sleep(1)
        driver.execute_script("window.scrollBy(0,1000)")
        # Scraping the Car Name
        for i in driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a"):
            name.append(i.text)
        # Scraping the Car Price value
        for i in driver.find_elements_by_xpath("//div[@class=' price']/span[2]"):
            price.append(i.text)
        # Scraping the Kilometer data    
        for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist-2']/div[1]"):
            running_km.append(i.text)
        # Scraping the Fuel Type    
        for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist-2']/div[2]"):
            fuel_type.append(i.text)
        # Scraping the Gear Transmission data    
        for i in driver.find_elements_by_xpath("//div[@class='truncate dotlist-2']/div[3]"):
            transmission.append(i.text)
        
```

```python
# Scrapping the car url
car_url=[]
for _ in range(110):
        time.sleep(1)
        driver.execute_script("window.scrollBy(0,1000)")
        current_url = driver.current_url    
        car = driver.find_elements_by_xpath("//div[@class='gsc_col-xs-7  carsName']/a")
        for i in car:
            car_url.append(i.get_attribute('href'))
```

```python
milleage=[]
color=[]
max_power=[]
engine=[]
seat=[]
front_brake=[]
length=[]
width=[]
height=[]
weight=[]
wheel_base=[]
front_tread=[]
rear_tread=[]
engine_type=[]

for i in car_url:
    driver.get(i)
    time.sleep(3)
    # Scraping the Car Milleage Data
    try:
        view_more = driver.find_element_by_xpath("//*[text() = 'View All Specifications' or text() = 'View More']")
        driver.execute_script("arguments[0].scrollIntoView();", view_more)
        driver.execute_script("arguments[0].click();", view_more)
        time.sleep(2)
    except NoSuchElementException:
        pass
    
    #Scraping the Milleage data
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Mileage']/following-sibling::div"):
            milleage.append(i.text.replace('kmpl',''))
    except NoSuchElementException:
        milleage.append('-')
        
    #Scraping the color:
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Color']/following-sibling::div"):
            color.append(i.text)
    except NoSuchElementException:
        color.append('-')
        
    #Scraping the Max Power:
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Max Power']/following-sibling::div"):
            max_power.append(i.text)
    except NoSuchElementException:
        max_power.append('-')
        
    #Scraping the Engine
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Engine']/following-sibling::div"):
            engine.append(i.text)
    except NoSuchElementException:
        engine.append('-')
        
    #Scraping the Seat
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Seats']/following-sibling::div"):
            seat.append(i.text)
    except NoSuchElementException:
        seat.append('-')
    
    #Scraping the Front Brake:
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Front Brake Type']/following-sibling::div"):
            front_brake.append(i.text)
    except NoSuchElementException:
        front_brake.append('-')
    
    #Scraping the Length
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Length']/following-sibling::div"):
            length.append(i.text.replace('mm',''))
    except NoSuchElementException:
        length.append('-')
        
    #Scraping the Width:
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Width']/following-sibling::div"):
            width.append(i.text.replace('mm',''))
    except NoSuchElementException:
        width.append('-')
        
    #Scraping the Height:
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Height']/following-sibling::div"):
            height.append(i.text.replace('mm',''))
    except NoSuchElementException:
        height.append('-')
        
    #Scraping the Kerb Weight:
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Kerb Weight']/following-sibling::div"):
            weight.append(i.text.replace('kg',''))
    except NoSuchElementException:
        weight.append('-')
        
    #Scraping the Wheel Base
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Wheel Base']/following-sibling::div"):
            wheel_base.append(i.text.replace('mm',''))
    except NoSuchElementException:
        wheel_base.append('-')
        
    #Scraping the Front Tread
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Front Tread']/following-sibling::div"):
            front_tread.append(i.text.replace('mm',''))
    except NoSuchElementException:
        front_tread.append('-')
        
    #Scraping the Rear Tread
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Rear Tread']/following-sibling::div"):
            rear_tread.append(i.text.replace('mm',''))
    except NoSuchElementException:
        rear_tread.append('-')
        
    #Scraping the Engine Type
    try:
        for i in driver.find_elements_by_xpath("//*[text()='Engine Type']/following-sibling::div"):
            engine_type.append(i.text)
    except NoSuchElementException:
        engine_type.append('-')
```
